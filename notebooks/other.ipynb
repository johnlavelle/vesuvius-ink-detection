```python
import torch
from torch.utils.data import Dataset, DataLoader
from collections import namedtuple

# Define the namedtuple
Sample = namedtuple('Sample', ['data', 'target'])

# Create a custom dataset class
class NamedTupleDataset(Dataset):
    def __init__(self, data, targets):
        self.data = data
        self.targets = targets

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        sample = Sample(self.data[idx], self.targets[idx])
        return sample

# Generate some random data
data = torch.randn(100, 3, 32, 32)  # 100 samples, 3 channels, 32x32 images
targets = torch.randint(0, 10, (100,))  # 100 random target values between 0 and 9

# Instantiate the custom dataset
dataset = NamedTupleDataset(data, targets)

# Create a DataLoader
dataloader = DataLoader(dataset, batch_size=10, shuffle=True, num_workers=2)

# Iterate through the DataLoader
for batch in dataloader:
    print(batch.data.shape)  # Access the data field
    print(batch.target.shape)  # Access the target field
```

```python
import numpy as np
import xarray as xr
import pandas as pd

# Create the xarray DataArray
data = np.array([[20, 22, 24, 26],
                 [21, 23, 25, 27],
                 [22, 24, 26, 28],
                 [23, 25, 27, 29]])

times = pd.date_range('2021-01-01', periods=4, freq='D')
locations = ['loc1', 'loc2', 'loc3', 'loc4']

temp = xr.DataArray(data, coords=[times, locations], dims=['time', 'location'])

# Apply rolling and construct
window_size = 3
rolling_temp = temp.rolling(time=window_size)
constructed_temp = rolling_temp.construct('window_dim')

# Calculate rolling mean
rolling_mean = constructed_temp.mean(dim='window_dim')

print(rolling_mean)
```

To find the rolling window size (w) as a function of length (L), the number of windows (n), and step size (s), you can use the following formula:

w = (L - (n - 1) * s) / n

```python
def win_size(L, n):
    return (2 * L) / (2 * n + 1)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader, ConcatDataset

# Define two datasets
class MyDataset1(Dataset):
    def __init__(self):
        self.data = [i**2 for i in range(10)]

    def __getitem__(self, index):
        if index >= len(self.data):
            raise IndexError("Index out of bounds")
        return self.data[index]

    def __len__(self):
        return len(self.data)

class MyDataset2(Dataset):
    def __init__(self):
        self.data = [i**3 for i in range(100)]

    def __getitem__(self, index):
        if index >= len(self.data):
            raise IndexError("Index out of bounds")
        return self.data[index]

    def __len__(self):
        return len(self.data)

# Concatenate the datasets
concat_dataset = ConcatDataset([MyDataset1(), MyDataset2()])

# Custom collate function to handle IndexError and return None for empty batches
def custom_collate(batch):
    filtered_batch = []
    for item in batch:
        if isinstance(item, Exception):
            print(f"Skipping item due to error: {item}")
            continue
        filtered_batch.append(item)

    if len(filtered_batch) == 0:
        return None
    return torch.tensor(filtered_batch)

# Set up the DataLoader with the custom collate function
dataloader = DataLoader(concat_dataset, batch_size=4, shuffle=True, collate_fn=custom_collate)

# Iterate over the DataLoader and check for None (empty batches)
for batch in dataloader:
    if batch is None:
        print("Skipping empty batch")
        continue
    print(batch)
    # Process the batch (e.g., forward pass, calculate loss, backpropagation, etc.)
```

```python
import torch
from collections import namedtuple
from torch.utils.data import Dataset, DataLoader

# Define the named tuple
MyNamedTuple = namedtuple('MyNamedTuple', ['tensor1', 'tensor2', 'tensor3'])

class MyDataset(Dataset):
    def __init__(self):
        self.data = [
            MyNamedTuple(
                torch.randn(1, 2),
                torch.randn(1),
                torch.randn(1, 3, 4, 5)
            ),
            MyNamedTuple(
                torch.randn(1, 2),
                torch.randn(1),
                torch.randn(1, 3, 4, 5)
            ),
            MyNamedTuple(
                torch.randn(1, 2),
                torch.randn(1),
                torch.randn(1, 3, 4, 5)
            )
        ]

    def __getitem__(self, index):
        return self.data[index]

    def __len__(self):
        return len(self.data)

# Custom collate function to handle stacking the tensors in the named tuple
def custom_collate(batch):
    stacked_tensor1 = torch.stack([nt.tensor1 for nt in batch])
    stacked_tensor2 = torch.stack([nt.tensor2 for nt in batch])
    stacked_tensor3 = torch.stack([nt.tensor3 for nt in batch])

    return MyNamedTuple(stacked_tensor1, stacked_tensor2, stacked_tensor3)

# Set up the DataLoader with the custom collate function
my_dataset = MyDataset()
dataloader = DataLoader(my_dataset, batch_size=3, shuffle=True, collate_fn=custom_collate)

# Iterate over the DataLoader and process the batches
for batch in dataloader:
    print(batch)
    # Process the batch (e.g., forward pass, calculate loss, backpropagation, etc.)
```

```python
import pandas as pd
from shapely.geometry import Point, Polygon
import random
import timeit

def check_points_in_polygon(points, polygon_coords):
    polygon = Polygon(polygon_coords)
    points_df = pd.DataFrame(points, columns=['x', 'y'])
    points_df['geometry'] = points_df.apply(lambda row: Point(row['x'], row['y']), axis=1)
    points_df['inside_polygon'] = points_df['geometry'].apply(lambda point: point.within(polygon))
    return points_df

points = [(random.uniform(-100, 100), random.uniform(-100, 100)) for _ in range(10000)]
polygon_coords = [(-50, -50), (-50, 50), (50, 50), (50, -50)]

execution_time = timeit.timeit("check_points_in_polygon(points, polygon_coords)", globals=globals(), number=1)
print(f"Execution time for 10,000 points: {execution_time:.4f} seconds")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.tensorboard import SummaryWriter
import numpy as np

# Create a simple model
class SimpleModel(nn.Module):
    def __init__(self):
        super(SimpleModel, self).__init__()
        self.linear = nn.Linear(1, 1)

    def forward(self, x):
        return self.linear(x)

# Generate random data
def generate_random_data(size):
    x = np.random.rand(size, 1).astype(np.float32)
    y = 3 * x + np.random.normal(0, 0.1, size=(size, 1)).astype(np.float32)
    return x, y

# Set up TensorBoard writer
writer = SummaryWriter("/home/john/code/kaggle/vesuvius/output/runs/")

# Set up model, loss function, and optimizer
model = SimpleModel()
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Train the model
num_epochs = 100
batch_size = 10

for epoch in range(num_epochs):
    x_train, y_train = generate_random_data(batch_size)
    x_train = torch.from_numpy(x_train)
    y_train = torch.from_numpy(y_train)

    # Forward pass
    outputs = model(x_train)
    loss = criterion(outputs, y_train)

    # Backward pass
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Log the loss to TensorBoard
    writer.add_scalar("Loss/train", loss.item(), epoch)

# Close the TensorBoard writer
writer.close()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class HybridModel1(nn.Module):
    def __init__(self, N):
        super(HybridModel1, self).__init__()
        # CNN part for vector input
        self.conv1 = nn.Conv2d(N, 16, 3, 1, 1)
        self.pool1 = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(16, 32, 3, 1, 1)
        self.pool2 = nn.MaxPool2d(2, 2)
        self.conv3 = nn.Conv2d(32, 64, 3, 1, 1)
        self.pool3 = nn.AdaptiveMaxPool2d((8, 8))
        self.flatten = nn.Flatten(start_dim=1)

        # FCN part for scalar input
        self.fc_scalar = nn.Linear(1, 16)

        # Combined layers (initialized later)
        self.fc_combined1 = None
        self.relu = nn.ReLU()
        self.fc_combined2 = nn.Linear(128, 1)  # Update this line
        self.sigmoid = nn.Sigmoid()

    def forward(self, x: torch.Tensor, scalar_input: torch.Tensor):
        # CNN part
        x = self.pool1(F.relu(self.conv1(x)))
        x = self.pool2(F.relu(self.conv2(x)))
        x = self.pool3(F.relu(self.conv3(x)))
        x = self.flatten(x)

        # FCN part
        scalar_out = F.relu(self.fc_scalar(scalar_input))

        # Initialize combined layers based on input depth
        if self.fc_combined1 is None:
            input_depth = x.shape[1]
            self.fc_combined1 = nn.Linear(64 * 8 * 8 + 16, 128).to(x.device)

        # Combine CNN and FCN outputs
        combined = torch.cat((x, scalar_out), dim=1)

        # Combined layers
        x = self.relu(self.fc_combined1(combined))
        x = self.sigmoid(self.fc_combined2(x))

        return x

# Create an instance of the modified HybridModel1 with N=6
model = HybridModel1(N=6)
print(model)
# Generate dummy data
vector_input = torch.randn(32, 6, 61, 61)  # batch_size=32, channels=6, width=61, height=61
scalar_input = torch.randn(32, 1)          # batch_size=32, channels=1

# Pass the dummy data through the model
output = model(vector_input, scalar_input)

# Print the output shape
print(output.shape)  # Should print: torch.Size([32, 1])
```
